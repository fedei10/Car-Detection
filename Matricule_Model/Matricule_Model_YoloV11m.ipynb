```python

from ultralytics import YOLO

model = YOLO("yolo11m.pt")
model.train(data='/content/data-science-1/data.yaml', epochs=50, batch=16, imgsz=640)
```

```python
import cv2
import matplotlib.pyplot as plt

# Perform inference on the image
results = model("/content/data-science-1/test/images/1198_jpg.rf.4597925dc3eff8e48a3c4f3e5c705a74.jpg")

# Display the results
for result in results:  # Iterate through the list
    annotated_image = result.plot()  # Use the 'plot()' method to get the annotated image

# Display the image using Matplotlib
plt.figure(figsize=(12, 8))
plt.imshow(cv2.cvtColor(annotated_image, cv2.COLOR_BGR2RGB))
plt.axis('off')
plt.show()
```
