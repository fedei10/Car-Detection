# Testing the Model with text Extraction Using EasyOCR

```python
!pip install ultralytics
!pip install opencv-python
!pip install easyocr
```

```python
import cv2
from ultralytics import YOLO
import matplotlib.pyplot as plt
import requests
import re
import easyocr

# Initialize EasyOCR reader
reader = easyocr.Reader(['en', 'ar'], gpu=True)

# Download fine-tuned model from github repository
model_url = "https://github.com/fedei10/Car-Detection/raw/main/Matricule_Model/Matricule_Yolov11m.pt"
model_path = "/content/Matricule_Yolov11m.pt"
response = requests.get(model_url, stream=True)
if response.status_code == 200:
    with open(model_path, "wb") as f:
        for chunk in response.iter_content(chunk_size=8192):
            f.write(chunk)
    print(f"Model downloaded successfully and saved to {model_path}")
else:
    print(f"Failed to download the model. Status code: {response.status_code}")

# Load the YOLO model
model = YOLO(model_path)

# Perform inference on a sample image
image_path = "/content/data-science-1/valid/images/1121_jpg.rf.dbf3b4efa9898912eb5979092e1c75b2.jpg"
results = model(image_path)

# Extract detected bounding boxes and isolate the "matricule" class
image = cv2.imread(image_path)

# Log detected classes
def log_detected_classes(results):
    for result in results:
        for box in result.boxes:
            print(f"Detected class: {int(box.cls)}, Confidence: {box.conf[0]:.2f}")

log_detected_classes(results)

def read_text_with_easyocr(roi):
    # Use EasyOCR to extract text
    result = reader.readtext(roi, detail=0)
    r =' '.join(result)
    # Split the text into left and right numbers
    cleaned_text = re.findall(r'\d+', r)
    if len(cleaned_text) >= 2:
        left_numbers = cleaned_text[1]
        right_numbers = cleaned_text[0]
    else:
        left_numbers = "Not Found"
        right_numbers = "Not Found"

    plate = left_numbers+" TUNIS "+right_numbers

    return plate

for result in results:
    # Iterate through detected objects
    for box in result.boxes:
        # Filter by class index (matricule == 0)
        if int(box.cls[0]) == 0:  # Adjust this index based on the actual YOLO model's class mapping
            # Extract bounding box coordinates
            x1, y1, x2, y2 = map(int, box.xyxy[0].tolist())
            # Crop the detected region (plate)
            roi = image[y1:y2, x1:x2]
            # Extract text directly from the cropped image using EasyOCR
            detected_text = read_text_with_easyocr(roi)
            # Annotate the image with bounding boxes and detected text
            cv2.rectangle(image, (x1, y1), (x2, y2), (0, 255, 0), 2)
            cv2.putText(image, f"{detected_text}", (x1, y1 - 10), cv2.FONT_HERSHEY_SIMPLEX, 0.9, (36, 255, 12), 2)

# Display the annotated image
plt.figure(figsize=(12, 8))
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.axis('off')
plt.show()

# Optionally save the annotated image
cv2.imwrite("annotated_image.jpg", image)
```
