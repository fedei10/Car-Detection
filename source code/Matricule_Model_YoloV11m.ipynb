### EasyOCR Workflow for License Plate Detection and Text Extraction

1. **Image Input**:
   - Read the input image using OpenCV.
   - `image = cv2.imread(image_path)`

2. **Vehicle Detection (YOLO)**:
   - Detect vehicles in the image using the vehicle YOLO model.
   - For each detected vehicle, extract the Region of Interest (ROI):
     ```python
     vehicle_roi = image[y1:y2, x1:x2]
     ```

3. **License Plate Detection (YOLO)**:
   - Detect license plates **inside** the vehicle ROI using the plate YOLO model.
   - Adjust coordinates of the license plate relative to the original image:
     ```python
     plate_x1 = x1 + px1
     plate_y1 = y1 + py1
     plate_x2 = x1 + px2
     plate_y2 = y1 + py2
     ```

4. **Extract License Plate ROI**:
   - Crop the license plate area:
     ```python
     plate_roi = image[plate_y1:plate_y2, plate_x1:plate_x2]
     ```

5. **Text Recognition with EasyOCR**:
   - Preprocess the cropped license plate:
     - Convert to **grayscale**.
     - Enhance contrast with **histogram equalization**.
     - Apply **binary thresholding** to segment text.
   - Perform OCR using EasyOCR:
     ```python
     result = reader.readtext(plate_roi, detail=0)
     ```

6. **Clean OCR Output**:
   - Remove Arabic numerals (`[\u0660-\u0669]`).
   - Extract numeric sequences:
     ```python
     cleaned_text = re.findall(r'\d+', result)
     ```

7. **Format the Text**:
   - Reconstruct the text in the format:
     - **Two sequences**: `{right_numbers} ØªÙˆÙ†Ø³ {left_numbers}`
     - **One sequence**: `Ù†.Øª{left_numbers}`
     ```python
     plate = f"{right_numbers} ØªÙˆÙ†Ø³ {left_numbers}"
     ```

8. **Black Pixel Percentage Check**:
   - Calculate the percentage of black pixels in the license plate ROI to validate it:
     ```python
     black_percentage = check_black_percentage(plate_roi)
     ```

9. **Annotate the Image**:
   - Overlay the detected text on the image using Pillow (`annotate_with_arabic_text`).
   - Draw bounding boxes for the vehicle and license plate using OpenCV.

10. **Save and Display**:
    - Save the annotated image to a file.
    - Display the image using Matplotlib.

**OCR Functions**

````python
# Log detected classes
def log_detected_classes(results):
    for result in results:
        for box in result.boxes:
            print(f"Detected class: {int(box.cls)}, Confidence: {box.conf[0]:.2f}")

# Extract text from the plate using EasyOCR
def read_text_with_easyocr(roi):
    result = reader.readtext(roi, detail=0)
    r = ' '.join(result)
    print(f"unclean text: {r}")
    r = re.sub(r'[\u0660-\u0669]', '', r)
    cleaned_text = re.findall(r'\d+', r)
    plate = ""
    print(f"cleaned text {cleaned_text}")
    if len(cleaned_text) == 2:
        left_numbers = cleaned_text[1]
        right_numbers = cleaned_text[0]
        plate = f"{right_numbers} ØªÙˆÙ†Ø³ {left_numbers}"
    elif len(cleaned_text) == 1:
        left_numbers = cleaned_text[0]
        plate = f"Ù†.Øª"+left_numbers
    return plate
from PIL import Image, ImageDraw, ImageFont
def annotate_with_arabic_text(image, text, position, font_path="/content/Car-Detection-main/Matricule_Model/Traditional Arabic Regular.ttf", font_size=40, color=(0, 255, 0)):
    """
    Annotates the image with Arabic text using Pillow.
    """
    # Convert OpenCV image to a Pillow image
    pil_image = Image.fromarray(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
    draw = ImageDraw.Draw(pil_image)

    # Load the font
    font = ImageFont.truetype(font_path, font_size)

    # Add text to the image
    draw.text(position, text, font=font, fill=color)

    # Convert the Pillow image back to OpenCV format
    annotated_image = cv2.cvtColor(np.array(pil_image), cv2.COLOR_RGB2BGR)
    return annotated_image

# Calculate the percentage of black pixels in the plate region
def check_black_percentage(roi):
    gray = cv2.cvtColor(roi, cv2.COLOR_BGR2GRAY)
    _, thresh = cv2.threshold(gray, 50, 255, cv2.THRESH_BINARY_INV)
    black_pixels = np.sum(thresh == 255)
    total_pixels = roi.shape[0] * roi.shape[1]
    black_percentage = (black_pixels / total_pixels) * 100
    return black_percentage
````

**Initialisation Yolo Model, EasyOCR and Imports**

````python
import cv2
from ultralytics import YOLO
import matplotlib.pyplot as plt
import requests
import re
import easyocr
import numpy as np

# Initialize EasyOCR reader
reader = easyocr.Reader(['en', 'ar'], gpu=True)

model_path = "/content/Car-Detection-main/Matricule_Model/Matricule_Yolov11m.pt"

# Load the YOLO model
plate_model = YOLO(model_path)
vehicle_model = YOLO("yolo11x.pt")
````

**Testing with images**

````python

# List of image paths to process
image_paths = [
    "/content/data-science-1/valid/images/291_jpg.rf.93210f4b8b1aaf3a612d70e85e0d6864.jpg",
    "/content/data-science-1/valid/images/1162_jpg.rf.2e18fea23e4395550869d4baa8832c78.jpg",
    "/content/462571617_2015709385594831_7943216827812566330_n.jpg"
]

# Process each image
for image_path in image_paths:
    print(f"Processing image: {image_path}")
    image = cv2.imread(image_path)  # Load the image

    # Step 1: Detect vehicles
    vehicle_results = vehicle_model(image)
    for vehicle_result in vehicle_results:
        for box in vehicle_result.boxes:
            if int(box.cls[0]) == 2:  # Class index for vehicles (e.g., "car" class in YOLOv8)
                x1, y1, x2, y2 = map(int, box.xyxy[0].tolist())

                # Extract vehicle ROI
                vehicle_roi = image[y1:y2, x1:x2]

                # Step 2: Detect license plates inside the vehicle ROI
                plate_results = plate_model(vehicle_roi)
                for plate_result in plate_results:
                    for plate_box in plate_result.boxes:
                        if int(plate_box.cls[0]) == 0:  # Class index for license plates
                            px1, py1, px2, py2 = map(int, plate_box.xyxy[0].tolist())

                            # Adjust license plate coordinates relative to the original image
                            plate_x1 = x1 + px1
                            plate_y1 = y1 + py1
                            plate_x2 = x1 + px2
                            plate_y2 = y1 + py2

                            # Extract license plate ROI
                            plate_roi = image[plate_y1:plate_y2, plate_x1:plate_x2]

                            # Step 3: OCR to read text (using your existing function)
                            detected_text = read_text_with_easyocr(plate_roi)
                            black_percentage = check_black_percentage(plate_roi)

                            print(f"Detected License Plate: {detected_text}")
                            print(f"Black Pixel Percentage: {black_percentage:.2f}%")

                            # Step 4: Annotate image if black percentage >= 47%
                            if black_percentage >= 47:
                                position = (plate_x1, plate_y1 - 50)
                                image = annotate_with_arabic_text(image, detected_text, position)

                                # Draw bounding box for the license plate
                                cv2.rectangle(image, (plate_x1, plate_y1), (plate_x2, plate_y2), (0, 255, 0), 2)

                # Draw bounding box for the vehicle
                cv2.rectangle(image, (x1, y1), (x2, y2), (255, 0, 0), 2)

    # Display the annotated image
    plt.figure(figsize=(12, 8))
    plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
    plt.axis('off')
    plt.show()

    # Save the annotated image with a unique filename
    output_filename = f"annotated_{os.path.basename(image_path)}"
    cv2.imwrite(output_filename, image)
    print(f"Saved annotated image as {output_filename}")
````

````python
from google.colab.patches import cv2_imshow
# Input and output video paths
video_path = "/content/Car-Detection-main/models/video_2.mp4"
output_video_path = "output_video.mp4"

# Read the input video
cap = cv2.VideoCapture(video_path)

# Video properties
fps = int(cap.get(cv2.CAP_PROP_FPS))
frame_width = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
frame_height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))
fourcc = cv2.VideoWriter_fourcc(*'mp4v')  # Codec for output video

# Create a VideoWriter object
out = cv2.VideoWriter(output_video_path, fourcc, fps, (frame_width, frame_height))

# Initialize frame counter
frame_count = 0
frame_skip = 30  # Number of frames to skip

# Process video frames
while cap.isOpened():
    ret, frame = cap.read()
    if not ret:
        break  # Stop when video ends

    # Only process every 'frame_skip' frames
    if frame_count % frame_skip == 0:
        print(f"Processing frame: {frame_count}")

        # Step 1: Detect vehicles
        vehicle_results = vehicle_model(frame)
        for vehicle_result in vehicle_results:
            for box in vehicle_result.boxes:
                if int(box.cls[0]) == 2:  # Class index for vehicles
                    x1, y1, x2, y2 = map(int, box.xyxy[0].tolist())

                    # Extract vehicle ROI
                    vehicle_roi = frame[y1:y2, x1:x2]

                    # Step 2: Detect license plates inside the vehicle ROI
                    plate_results = plate_model(vehicle_roi)
                    for plate_result in plate_results:
                        for plate_box in plate_result.boxes:
                            if int(plate_box.cls[0]) == 0:  # Class index for license plates
                                px1, py1, px2, py2 = map(int, plate_box.xyxy[0].tolist())

                                # Adjust license plate coordinates relative to the original frame
                                plate_x1 = x1 + px1
                                plate_y1 = y1 + py1
                                plate_x2 = x1 + px2
                                plate_y2 = y1 + py2

                                # Extract license plate ROI
                                plate_roi = frame[plate_y1:plate_y2, plate_x1:plate_x2]

                                # Step 3: OCR to read text
                                detected_text = read_text_with_easyocr(plate_roi)
                                black_percentage = check_black_percentage(plate_roi)

                                print(f"Detected License Plate: {detected_text}")
                                print(f"Black Pixel Percentage: {black_percentage:.2f}%")

                                # Annotate frame if black percentage >= 47%

                                position = (plate_x1, plate_y1 - 10)
                                frame = annotate_with_arabic_text(frame, detected_text, position)

                                # Draw bounding box for the license plate
                                cv2.rectangle(frame, (plate_x1, plate_y1), (plate_x2, plate_y2), (0, 255, 0), 2)

                    # Draw bounding box for the vehicle
                    cv2.rectangle(frame, (x1, y1), (x2, y2), (255, 0, 0), 2)

        # Write the annotated frame to the output video
        out.write(frame)

        # Display the frame (optional)
        cv2_imshow(frame)
        if cv2.waitKey(1) & 0xFF == ord('q'):  # Press 'q' to stop processing early
            break

    # Increment the frame counter
    frame_count += 1

# Release video objects
cap.release()
out.release()
cv2.destroyAllWindows()

print(f"Annotated video saved as: {output_video_path}")
````
